```python
# ------- Competition Functions ---------
def calc_accel_score(tMin, tYour):
    tMax = tMin * 1.5
    if tYour > tMax:
        return 4.5
    return round(((tMax/tYour)-1) / ((tMax/tMin) - 1) * 95.5 + 4.5, 2)

def calc_skidpad_score(tMin, tYour, cones):
    tMax = tMin * 1.25
    tYour = tYour + 0.125*cones
    if tYour > tMax:
        return 3.5
    return round(((tMax/tYour)**2-1) / ((tMax/tMin)**2 - 1) * 71.5 + 3.5, 2)

def calc_autox_score(tMin, tYour, cones, offtrack):
    tMax = tMin * 1.45
    tYour = tYour + 2*cones + 20*offtrack
    if tYour > tMax:
        return 6.5
    return round(((tMax/tYour) - 1) / ((tMax/tMin) - 1) * 118.5 + 6.5, 2)

def calc_endurance_score(tMin, tYour, cones, offtrack, lapTotal, lapYour):
    tMax = tMin * 1.45
    tYour = tYour + 2*cones + 20*offtrack
    score = round(((tMax/tYour) - 1) / ((tMax/tMin) - 1) * 250, 2)
    return score + (25 * lapYour/lapTotal)

def calc_efficiency_factor(tMin, tYour, co2Min, co2Your, lapTotal, lapYour):
    return ((tMin/lapTotal)/(tYour/lapYour)) * ((co2Min/lapTotal)/(co2Your/lapYour))

def calc_efficiency_score(tMin, tYour, co2Min, co2Your, lapTotal, lapYour):
    ef_min = calc_efficiency_factor(tMin, tMin * 1.45, co2Min, 60.06, lapTotal, lapTotal)
    ef_max = 0.784
    ef_your = calc_efficiency_factor(tMin, tYour, co2Min, co2Your, lapTotal, lapYour)
    return 100 * (ef_your - ef_min)/(ef_max - ef_min)
    
```

```python
%matplotlib inline
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import math
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib.collections import LineCollection
import os

# --- File import/clean Functions
def read_eng_file(file_path_eng, skip_row = 3):
    df_eng = pd.read_csv(file_path_eng, encoding='ISO-8859-1', sep='\t', skiprows=skip_row, low_memory=False)
    key_log_cols = ['Time', 'RPM', 'VSS1', 'TPS', 'MAP', 'AFR', 'CLT', 'MAT', 'Oil Temp Raw', 'Rad Temp', 'Oil Pressure', 'Fuel Pressure', 'Power']
    for col in key_log_cols:
        df_eng[col] = pd.to_numeric(df_eng[col], errors='coerce')
    return df_eng

def read_dyno_file(file_path_dyno):
    df = pd.read_csv(file_path_dyno, encoding='ISO-8859-1', skiprows=6, low_memory=False)
    #Clean Dyno Data File
    df.drop(df.columns[0], axis=1, inplace=True)
    df.drop(df.columns[3], axis=1, inplace=True)
    df['hp'] = pd.to_numeric(df['hp'], errors='coerce')
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df = df.dropna(subset=['s', 'mph', 'hp'])

    # Find the index where the value first goes negative and trim
    neg_index = df[df['hp'] < 0].index.min()
    df = df.iloc[:df.index.get_loc(neg_index)]
        
    return df

def calc_rpm_mph(speed_mph, gear, fdr, wr):
    speed_mps = speed_mph * 0.44704  # Convert mph to m/s
    wheel_rpm = (speed_mps * 60) / (2 * np.pi * wr)
    engine_rpm = wheel_rpm * gear_ratios[gear] * fdr * r_p
    return engine_rpm

def calc_dyno_torque(df, gear, fdr, wr):
    
    
    df['rpm'] = calc_rpm_mph(df['mph'], gear, fdr, wr)
    conversion_ratio = 5252 * 1.3558179483
    df['torque_Nm'] = (conversion_ratio * df['hp']) / df['rpm']

    # --- Add Linear interp
    '''
    start = 0.1
    num_points = 100  # Number of points to generate — adjust as needed
    
    mph_new = np.linspace(start, df.iloc[0]['mph'], num_points, endpoint=False)
    hp_new = np.linspace(start, df.iloc[0]['hp'], num_points, endpoint=False)
    torque_new = np.linspace(4, df.iloc[0]['torque_Nm'], num_points, endpoint=False)
    
    df_pad = pd.DataFrame({
        's': np.nan,
        'mph': mph_new,
        'hp': hp_new,
        'torque_Nm': torque_new,
        'rpm': np.nan
    })
    df_pad['rpm'] = calc_rpm_mph(df_pad['mph'], gear, fdr, wr)
    df = pd.concat([df_pad, df], ignore_index=True)
    '''
    
    #df = dyno_df.dropna(thresh=10)
    df.reset_index(inplace=True)
    df = df.drop(columns=['index', 'mph'])
    return df
    
# ----------- Function Definitions -------------------
# Calculate Driveline Force (n)
def calculate_driveline_force(torque, primary_ratio, gear_ratio, final_drive_ratio, wheel_radius):
  return (torque * gear_ratio * final_drive_ratio * primary_ratio) / wheel_radius

# Calculate Vehicle Speed (m/s)
def calculate_vehicle_speed(rpm, wheel_radius, gear_ratio):
  return (rpm / 60 * wheel_radius * 2 * math.pi * 3.6 / (gear_ratio)) / 3.6

def calc_rpm(speed, gear, fdr, wr):
    wheel_rpm = (speed * 60) / (2 * np.pi * wr) # Calculate wheel RPM first (speed to rotational speed of wheel)
    engine_rpm = wheel_rpm * gear_ratios[gear] * fdr * r_p # Multiply by gear ratio and final drive ratio to get engine RPM
    return engine_rpm

def calc_speed(engine_rpm, gear, fdr, wr):
    wheel_rpm = engine_rpm / (gear_ratios[gear] * fdr * r_p) # Calculate wheel RPM from engine RPM
    speed = (wheel_rpm * 2 * np.pi * wr) / 60 # Convert wheel RPM to vehicle speed (m/s)
    return speed

def calc_drag_force(speed_mps, Cd, A, air_density):
    F_drag = 0.5 * air_density * Cd * A * speed_mps**2 # Drag Force (N)
    return F_drag

# --------- Simulation Loop -----------
def simulate_acceleration_pull(v, velocity_at_time_t):
        # --- Initialize arrays ---
        velocity = [0]  # Initial speed in m/s
        acceleration_mps2 = [0]  # To store calculated accelerations
        engine_rpm = [0] # To store calculated rpm
        weight_shift = [v.Wf_ratio]
        shift_times = []
        clutch_engagment = [0]
        accel_pull_time = 0
        distance_m = [0]  # Start at 0 m
        #driveline_force = [0]
        max_traction_force = [0]
        traction_delta = [0]
        shift_timer = 0
        velocity_at_time = 0


        gear = 1
        speed = v.gear_df[f'Vehicle Speed (m/s) (Gear {gear})']
        force = v.gear_df[f'Driveline Force (Gear {gear})']
        rpm = v.gear_df['rpm']

        # Initial normal forces (before acceleration)
        F_normal_rear = v.vMass * g * (1 - v.Wf_ratio)  # rear axle normal force (N)
        F_normal_front = v.vMass * g * v.Wf_ratio  # front axle normal force (N)

        # --- Simulation loop ---
        for t in time_array[1:]:
            current_speed = velocity[-1]  # Last speed in m/s

            # --- Handle gear shifting ---
            if gear < no_gear_opts and current_speed >= v.shift_points[gear-1] and distance_m[-1] < 65:
                shift_times.append(t)

                gear += 1  # Shift to the next gear
                speed = v.gear_df[f'Vehicle Speed (m/s) (Gear {gear})']
                force = v.gear_df[f'Driveline Force (Gear {gear})']

                shift_timer = v.shift_delay_s  # Start the shift delay timer

            rpm = calc_rpm(current_speed, gear, v.fdr, v.wheel_radius)


            # Calculate weight transfer
            if len(acceleration_mps2) > weight_shift_avg_steps:
                accel_avg = np.mean(acceleration_mps2[-weight_shift_avg_steps:])
            else:
                accel_avg = np.mean(acceleration_mps2)
            delta_F_rear = (accel_avg * v.vMass * v.h_cg) / L  # rear weight transfer (N

            # Updated normal forces after weight transfer
            F_normal_rear_new = F_normal_rear + delta_F_rear
            F_normal_front_new = F_normal_front - delta_F_rear  # Front axle loses the same amount of weight

            # Calculate new front weight ratio
            new_Wf_ratio = F_normal_front_new / (F_normal_front_new + F_normal_rear_new)

            # Traction force at the rear
            μ_effective = μ_max - k * (F_normal_rear_new - F_normal_rear)
            F_traction_max = μ_effective * F_normal_rear_new

            driveline_force_N = np.interp(current_speed, speed, force)

            # --- During shift delay, no acceleration ---
            clutch_eng = 1
            if shift_timer > 0:
                shift_timer -= time_step_s
            else:
                if rpm < launch_rpm:
                    driveline_force_N = np.interp(calc_speed(launch_rpm, gear, v.fdr, v.wheel_radius), speed, force)
                    clutch_eng = clutchFactor
                else:
                    clutch_eng = 0

            # Interpolate driveline force (N) from DataFrame based on current speed (m/s)
            driveline_force_N = driveline_force_N * (1-clutch_engagment[-1])
            final_road_force = min(driveline_force_N, F_traction_max)
            # Acceleration due to drag (negative, slows the vehicle down)
            drag = -calc_drag_force(current_speed, Cd, v.frontal_area, air_density) / v.vMass

            acceleration = (final_road_force / v.vMass) + drag

            # Update speed (Euler integration): v = v_prev + a * dt
            new_speed = current_speed + acceleration * time_step_s

            distance_m.append(distance_m[-1] + velocity[-1] * time_step_s)  # v * dt
            if (distance_m[-1] >= 75 or t >= total_time_s - time_step_s) and accel_pull_time < 0.1:
                accel_pull_time = t


            #driveline_force.append(driveline_force_N)
            max_traction_force.append(F_traction_max)
            acceleration_mps2.append(acceleration)
            engine_rpm.append(rpm)
            weight_shift.append(new_Wf_ratio)
            velocity.append(new_speed)
            clutch_engagment.append(clutch_eng)
            traction_delta.append(F_traction_max - driveline_force_N)

            if t >= velocity_at_time_t:
                velocity_at_time = new_speed
                break

        # --- End of Simulation ---
        if accel_pull_time < 0.1:
            accel_pull_time = total_time_s
        # --- Convert lists to arrays ---
        acceleration_mps2 = np.array(acceleration_mps2)
        velocity = np.array(velocity)
        distance_m = np.array(distance_m)
        engine_rpm = np.array(engine_rpm)
        weight_shift = np.array(weight_shift)
        shift_times = np.array(shift_times)
        clutch_engagment = np.array(clutch_engagment)
        velocity_at_time = np.array(velocity_at_time)
        accel_pull_time = np.array(accel_pull_time)
        #driveline_force = np.array(driveline_force)
        max_traction_force = np.array(max_traction_force)
        v.shift_points = np.array(v.shift_points)
        traction_delta = np.array(traction_delta)

        return velocity, acceleration_mps2, distance_m, engine_rpm, weight_shift, shift_times, clutch_engagment, accel_pull_time, traction_delta, max_traction_force, velocity_at_time

class Vehicle:
    def __init__(this, m, fwr, wr, p_mult, final_dr, st, cg_h, fa, df):
        this.vMass = m
        this.Wf_ratio = fwr
        this.wheel_radius = wr
        this.power_mult = p_mult
        this.fdr = final_dr
        this.shift_delay_s = st
        this.h_cg = cg_h
        this.frontal_area = fa; # frontal area [m^2]
        this.accel_pull_time = 0
        this.shift_p_best = []
        this.speed_gate_times = []
        this.df = df
        this.gear_df = pd.DataFrame()
        this.gear_df['rpm'] = df['rpm']

    def calc_shift_points(this):
        # Loop through each gear ratio
        for gear in gear_ratios:
            # Vehicle speed in m/s
            this.gear_df[f'Vehicle Speed (m/s) (Gear {gear})'] = calculate_vehicle_speed(this.df['rpm'], this.wheel_radius, gear_ratios[gear] * this.fdr * r_p)
            # Driveline force (N)
            this.gear_df[f'Driveline Force (Gear {gear})'] = this.df.apply(
                lambda row: calculate_driveline_force(row['torque_Nm'] * this.power_mult, r_p, gear_ratios[gear], this.fdr, this.wheel_radius)
                if row['torque_Nm'] > 0 else 0,
                axis=1
            )

        # --- Calculate shift points between gears for this FDR ---
        shift_points = []

        for gear1, gear2 in gear_combinations:
            speed1 = this.gear_df[f'Vehicle Speed (m/s) (Gear {gear1})']
            force1 = this.gear_df[f'Driveline Force (Gear {gear1})']
            speed2 = this.gear_df[f'Vehicle Speed (m/s) (Gear {gear2})']
            force2 = this.gear_df[f'Driveline Force (Gear {gear2})']

            # Initialize min_diff for each pair
            min_diff = float('inf')
            min_diff_speed = 0

            if len(shift_points) > 0:
                start_idx = (abs(speed1 - shift_points[-1])).idxmin()
            else:
                start_idx = 0
            # Iterate through speed values
            for i in range(start_idx, len(speed1)):
                speed = speed1[i]

                # Find the closest speed in the other gear
                idx_closest = (abs(speed2 - speed)).idxmin()

                abs_diff = abs(force1[i] - force2[idx_closest])

                if abs_diff < min_diff:
                    min_diff = abs_diff
                    min_diff_speed = speed

            # Store the shift point speed for this gear pair
            shift_points.append(min_diff_speed)
        this.opt_shift_points = np.array(shift_points)
        this.shift_points = np.array(shift_points)
        
    def run_accel_sim(this):
        this.calc_shift_points()
        best_shift_points = None
        best_accel_pull_time = float('inf')
        best_velocity = 0

        # Calculate times in between shifts to optimize velocity for
        velocity, acceleration_mps2, distance_m, engine_rpm, weight_shift, shift_times, clutch_engagment, accel_pull_time, traction_delta, max_traction_force, velocity_at_t = simulate_acceleration_pull(v, total_time_s)
        this.speed_gate_times = []
        for i in range(len(shift_times) - 1):
            this.speed_gate_times.append(((shift_times[i + 1] - shift_times[i] - this.shift_delay_s) / 2) + shift_times[i] + this.shift_delay_s)
        
        this.new_shift_points = this.opt_shift_points.copy()
        for gear_num in range(len(gear_ratios)-2):
          if gear_num < len(this.speed_gate_times):
            for c, a in enumerate(sp_adj_array):
                this.shift_points = this.new_shift_points.copy()
                this.shift_points[gear_num] = this.shift_points[gear_num] + a
                
                velocity, acceleration_mps2, distance_m, engine_rpm, weight_shift, shift_times, clutch_engagment, accel_pull_time, traction_delta, max_traction_force, speed_gate_velocity = simulate_acceleration_pull(v, this.speed_gate_times[gear_num])
                #print(f'adj: {a:.2f} - shift times {this.shift_points} : speed: {speed_gate_velocity:.2f}')
                if best_shift_points is None or speed_gate_velocity > best_velocity:
                    #print("--- New Best ---")
                    best_velocity = speed_gate_velocity
                    best_shift_points = this.shift_points
            this.new_shift_points = best_shift_points
        this.shift_points = this.new_shift_points

        this.velocity, this.acceleration_mps2, this.distance_m, this.engine_rpm, this.weight_shift, this.shift_times, this.clutch_engagment, this.accel_pull_time, this.traction_delta, this.max_traction_force, this.speed_gate_velocity = simulate_acceleration_pull(v, total_time_s)
        #print(f'   --- Best time: {this.accel_pull_time:.2f}                 --- Shift Point adj: {this.shift_p_best}')
    def print_time(self):
        print(f"Best Time: {self.accel_pull_time:.3f}, Accel Score(2024): {calc_accel_score(comp_min_accel_t, self.accel_pull_time)}")
#End Vehicle Class





# ------ Graphs ----------
def plot_4_variables(df, df_column_names, x_range=None, y_range=None):
    if x_range is None:
        x_min = df[df_column_names[0]].min()
        x_max = df[df_column_names[0]].max()
        x_range = [x_min, x_max]

    if y_range is not(None):
        y_min = y_range[0]
        y_max = y_range[1]
    
    # Clip the DataFrame to the selected x_range
    df = df[(df[df_column_names[0]] >= x_range[0]) & (df[df_column_names[0]] <= x_range[1])].copy()
    
    fig, ax1 = plt.subplots(figsize=(14, 6))
    colors = ['tab:blue', 'tab:red', 'tab:green', 'tab:orange']

    # Set x-axis limits before plotting
    ax1.set_xlim(x_range[0], x_range[1])
    
    # Track all y-axes for label alignment
    axes = [ax1]
    
    # Clear legend and prepare label placement
    y_label_x_pos = 0.85  # X-position for all text labels (relative to axes)
    y_label_y_start = 0.98  # Start from top
    y_label_spacing = 0.04  # Vertical spacing between entries

    plt_title = df_column_names[0] + " vs "
    
    for i in range(len(df_column_names) - 1):
        column = df_column_names[i + 1]
        
        color = colors[i]
        df[column] = pd.to_numeric(df[column], errors='coerce')
    
        if i == 0:
            cur_graph = ax1
            plt_title += column
        else:
            plt_title += ", " + column
            cur_graph = ax1.twinx()
            axes.append(cur_graph)
            cur_graph.spines['right'].set_position(('outward', 60 * i))

        if y_range is None:
            y_min = df[column].min()
            y_max = df[column].max()
            y_range = [y_min, y_max]
        
        cur_graph.set_ylim(y_range)
    
        # Plot data
        cur_graph.plot(df[df_column_names[0]], df[column], color=color, linewidth=1)
        cur_graph.set_ylabel(column, color=color)
    
        # Add ymin/ymax text at top-right, stacked
        label = f"{column}: {y_min:.1f} - {y_max:.1f}"
        ax1.text(
            y_label_x_pos,              # x position in axes coords
            y_label_y_start - i * y_label_spacing,  # y position
            label,
            transform=ax1.transAxes,
            color=color,
            ha='left',
            va='top',
            fontsize=9
        )

    plt.title(plt_title)
    plt.tight_layout()
    plt.show()



def plot_graphs(v):
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 10))

    # --- Left y-axis (Acceleration) ---
    color_acc = 'tab:red'
    ax1.set_ylabel('Acceleration (m/s²)', color=color_acc)
    accel_plot = ax1.plot(time_array, v.acceleration_mps2, label='Acceleration (m/s²)', color=color_acc, linewidth=2.5)
    ax1.tick_params(axis='y', labelcolor=color_acc)
    accel_plot = ax1.plot(time_array, v.max_traction_force / v.vMass, label='Acceleration Max (m/s²)', color=color_acc, linewidth=1, linestyle='--')

    # --- Right y-axis (Speed) ---
    ax1_2 = ax1.twinx()
    color_speed = 'tab:blue'
    ax1_2.set_xlabel('Time (s)')
    ax1_2.set_ylabel('Vehicle Speed (km/h)', color=color_speed)
    speed_plot = ax1_2.plot(time_array, v.velocity * 3.6, label='Vehicle Speed (km/h)', color=color_speed, linewidth=2)
    ax1_2.tick_params(axis='y', labelcolor=color_speed)

    for gate_time in v.speed_gate_times:
      if gate_time < total_time_s:
        ax1.axvline(x=gate_time, color='orange', linestyle=':', linewidth=1.5)


    for i, shift_t in enumerate(v.shift_times):
        # Find the closest index to the shift time
        idx = (np.abs(time_array - shift_t)).argmin()

        # Get the RPM at that time
        rpm_at_shift = v.engine_rpm[idx-1]

        # Plot the vertical line
        ax1.axvline(x=shift_t, color='green', linestyle='--', linewidth=1.5)

        # Add a label with Shift number and RPM
        ax1.text(
            shift_t,
            ax1.get_ylim()[1] * 0.99,  # Adjust Y position as needed
            f'Shift ->{i+1}\n{rpm_at_shift:.0f} RPM',
            color='green',
            rotation=90,
            verticalalignment='top',
            horizontalalignment='right',
            fontsize=8,
            fontweight='bold'
        )

    # Vertical line at time where distance crosses 75 m
    ax1.axvline(x=v.accel_pull_time, color='purple', linestyle='-', linewidth=1.5)
    ax1.text(
          v.accel_pull_time,                             # x-coordinate (same as the line)
          ax1.get_ylim()[1] * 0.99,      # y-coordinate, near the top of the y-axis
          f'75 m\n{v.accel_pull_time:.2f}',        # Label text
          color='black',
          verticalalignment='top',       # Aligns text relative to the y value
          horizontalalignment='right',   # Aligns text relative to the x value
          fontsize=8,
          fontweight='bold')

    plt.title(f'Simulated Accel Pull (with {v.shift_delay_s}s Shift Delay, {v.fdr:.2f} FDR, {v.vMass} Vehicle Mass')
    ax1.grid(True)

    # Second graph

    # --- Left y-axis (Traction Delta) ---
    # Prepare points and segments for LineCollection
    points = np.array([time_array, v.traction_delta]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)

    # Define colors: blue if >=0, red if <0
    colors = ['tab:blue' if td >= 0 else 'red' for td in v.traction_delta[:-1]]

    # Create LineCollection
    lc = LineCollection(segments, colors=colors, linewidths=3)

    ax2.set_ylabel('Wheel Traction Delta (Traction max - Driveline Force)', color='tab:blue')
    ax2.add_collection(lc)

    # Set the limits for x and y axes based on your data
    ax2.set_xlim(time_array[0], time_array[-1])
    ax2.set_ylim(min(v.traction_delta) * 1.1, max(v.traction_delta) * 1.1)

    # --- Third y-axis (Distance) ---
    ax2_3 = ax2.twinx()

    color_weight_shift = 'tab:green'
    ax2_3.set_ylabel('Weight Shift (% front)', color=color_weight_shift)
    dist_plot = ax2_3.plot(time_array, v.weight_shift, label='Weight Shift', color=color_weight_shift, linewidth=1)
    ax2_3.tick_params(axis='y', labelcolor=color_weight_shift)

    # --- Title and Grid ---
    ax2.grid(True)

    drag = []
    for t in time_array:
        drag.append(calc_drag_force(v.velocity[np.where(time_array == t)[0][0]], Cd, v.frontal_area, air_density))

    # --- Left y-axis (Acceleration) ---
    color_acc = 'tab:red'
    ax3.set_ylabel('RPM', color=color_acc)
    accel_plot = ax3.plot(time_array, v.engine_rpm, label='RPM', color=color_acc, linewidth=2)
    ax3.tick_params(axis='y', labelcolor=color_acc)

    plt.show()

def plot_shift_points(v):   
    plt.figure(figsize=(12, 6))
    plt.title(f"Driveline Force vs. Vehicle Speed for Each Gear (FDR: {v.fdr})")

    for gear in gear_ratios:
        speed = v.gear_df[f'Vehicle Speed (m/s) (Gear {gear})'] * 3.6 # Convert m/s to km/h
        force = v.gear_df[f'Driveline Force (Gear {gear})']
        plt.plot(speed, force, label=f'Gear {gear}')

        if gear < 6:
            #Plot Optimal Shift Points
            shift_pt_speed = v.opt_shift_points[gear-1] * 3.6 # Convert m/s to km/h
            shift_pt_force = v.gear_df.loc[(speed - shift_pt_speed).abs().idxmin(), f'Driveline Force (Gear {gear})']
            plt.plot(shift_pt_speed, shift_pt_force, marker='o', color='red', markersize=8, alpha=0.5)

            if v.shift_points is not None:
                #Plot adjusted Shift Points
                shift_pt_speed = v.shift_points[gear-1] * 3.6 # Convert m/s to km/h
                shift_pt_force = v.gear_df.loc[(speed - shift_pt_speed).abs().idxmin(), f'Driveline Force (Gear {gear})']
                plt.plot(shift_pt_speed, shift_pt_force, marker='o', color='green', markersize=4, alpha=0.9)

    plt.xlabel("Vehicle Speed (km/h)")
    plt.ylabel("Driveline Force (N)")
    plt.grid(True)
    plt.legend()
    plt.show()

def plot_time_score_vs_iterator(var_iter_array, vehicles, iterator_name):

    plt.figure(figsize=(10, 6))

    # Primary y-axis (Acceleration Time)
    ax1 = plt.gca()
    line1, = ax1.plot(var_iter_array, [v.accel_pull_time for v in vehicles], marker='o', label="Acceleration Time", color='b')
    line2 = ax1.axhline(comp_min_accel_t * 1.5, color='green', linewidth=1.5, label='Comp Max Time')
    line3 = ax1.axhline(comp_min_accel_t, color='green', linewidth=1.5, label='Comp Min Time')

    # Secondary y-axis (Score)
    ax2 = ax1.twinx()
    # Calculate score changes (first value is just the score itself)
    scores = [calc_accel_score(comp_min_accel_t, v.accel_pull_time) for v in vehicles]
    score_changes = np.diff(scores, prepend=scores[0])  # First value stays the same

    # Plot the change in score
    line4, = ax2.plot(var_iter_array, score_changes, marker='s', markersize=3, linestyle='--', linewidth=1, color='r', label="Score Change")


    # 🏷️ Set Axis Labels
    ax1.set_xlabel(iterator_name, fontsize=12)  # X-axis label
    ax1.set_ylabel('Acceleration Time (s)', fontsize=12, color='b')  # Left Y-axis label
    ax2.set_ylabel('Change in Score', fontsize=12, color='r')  # Right Y-axis label

    # Grid, Title, and Legends
    plt.title(f'{iterator_name} vs Acceleration Time & Score', fontsize=14)
    ax1.grid(True)

    # Combine Legends from Both Axes
    lines = [line1, line2, line3, line4]
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper right', fontsize=10)

    print(f"Scores: {scores}")
    plt.show()

def print_vehicle_info(v):
    print(f":: VMass {v.vMass}, Front Weight Ratio {v.Wf_ratio}, Wheel Radius {v.wheel_radius}, FDR {v.fdr:.2f}, Shift Time {v.shift_delay_s}, CG Height {v.h_cg}, Frontal Area {v.frontal_area}")

# ------- Competition Functions ---------
def calc_accel_score(tMin, tYour):
    tMax = tMin * 1.5
    if tYour > tMax:
        return 4.5
    return round(((tMax/tYour)-1) / ((tMax/tMin) - 1) * 95.5 + 4.5, 2)

def calc_skidpad_score(tMin, tYour, cones):
    tMax = tMin * 1.25
    tYour = tYour + 0.125*cones
    if tYour > tMax:
        return 3.5
    return round(((tMax/tYour)**2-1) / ((tMax/tMin)**2 - 1) * 71.5 + 3.5, 2)

def calc_autox_score(tMin, tYour, cones, offtrack):
    tMax = tMin * 1.45
    tYour = tYour + 2*cones + 20*offtrack
    if tYour > tMax:
        return 6.5
    return round(((tMax/tYour) - 1) / ((tMax/tMin) - 1) * 118.5 + 6.5, 2)

def calc_endurance_score(tMin, tYour, cones, offtrack, lapTotal, lapYour):
    tMax = tMin * 1.45
    tYour = tYour + 2*cones + 20*offtrack
    score = round(((tMax/tYour) - 1) / ((tMax/tMin) - 1) * 250, 2)
    return score + (25 * lapYour/lapTotal)

def calc_efficiency_factor(tMin, tYour, co2Min, co2Your, lapTotal, lapYour):
    return ((tMin/lapTotal)/(tYour/lapYour)) * ((co2Min/lapTotal)/(co2Your/lapYour))

def calc_efficiency_score(tMin, tYour, co2Min, co2Your, lapTotal, lapYour):
    ef_min = calc_efficiency_factor(tMin, tMin * 1.45, co2Min, 60.06, lapTotal, lapTotal)
    ef_max = 0.784
    ef_your = calc_efficiency_factor(tMin, tYour, co2Min, co2Your, lapTotal, lapYour)
    return 100 * (ef_your - ef_min)/(ef_max - ef_min)
```
